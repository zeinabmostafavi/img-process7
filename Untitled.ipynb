```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
import random
import imageio
```

```python
img_tv=cv2.imread('tv.jpg')
img_tv=cv2.cvtColor(img_tv,cv2.COLOR_BGR2GRAY)
plt.imshow(img_tv,cmap='gray')
```

```python
roi = img_tv[20:200, 100:400]
plt.imshow(roi, cmap='gray')
```

```python
def snowflake(img_tv,mask):
    snflk = np.zeros(img_tv.shape,np.uint8)
    rvs = 1 - mask
    for i in range(img_tv.shape[0]):
        for j in range(img_tv.shape[1]):
            rand = random.random()
            if rand < mask:
                snflk[i][j] = 0
            elif rand > rvs:
                snflk[i][j] = 255
            else:
                snflk[i][j] = img_tv[i][j]
    return snflk
```

```python
img_snowflake = snowflake(roi, 0.10)
img_tv[20:200, 100:400] = img_snowflake
plt.imshow(img_tv, cmap='gray')
```

```python
images = []
for i in range(40):
    img_tv=cv2.imread('tv.jpg',cv2.IMREAD_GRAYSCALE)
    screen=img_tv[20:200,100:400]
    noise = snowflake(screen,0.10)
    img_tv[20:200,100:400]=noise
    images.append(img_tv)

imageio.mimwrite('snowflake0.gif', images)
```
